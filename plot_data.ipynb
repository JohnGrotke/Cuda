```python
import struct
import matplotlib.pyplot as plt

def read_binary_file(filename, var_type):
    with open(filename, 'rb') as file:
        binary_data = file.read()
    num_elements = len(binary_data) // struct.calcsize(var_type)
    data = struct.unpack('d' * num_elements, binary_data)
    return data

data = read_binary_file('vector_data.bin', 'd')

plt.plot(data)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Vector Plot')
plt.show()
```
